```python
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
tables = pd.read_html(url, storage_options={"User-Agent": "Mozilla/5.0"})
df = tables[0]
print(df.head())
```

```python
#symbol을 리스트로 변환해서 저장
symbols = df['Symbol'].tolist()
print(len(symbols))
print(symbols[:5])
```

```python
import yfinance as yf
import time

#회사에 맞는 정보 긁어옴 (야후 파이낸스에서)
ticker = yf.Ticker("AAPL")
#데이터 클리닝 (존재하지 않는 데이터 제거)
earnings = ticker.earnings_dates.dropna()
print(earnings)
```

```python
earnings_dates = []
for symbol in symbols:
    try:
        ticker = yf.Ticker(symbol)
        earnings = ticker.earnings_dates.reset_index().dropna()
        earnings['Symbol'] = symbol
        earnings_dates.append(earnings)
    except Exception as e:
        print(f"{symbol}: {e}")
    time.sleep(0.2)

all_earnings = pd.concat(earnings_dates)
print(all_earnings)

all_earnings.to_csv("earnings_data.csv", index=False)
```
